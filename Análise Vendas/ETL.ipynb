```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Importando Dados
tabela_clientes = pd.read_csv("1-datasets/Clientes.csv", sep=";")
tabela_pedidos = pd.read_csv("1-datasets/Pedidos.csv", sep=";")
tabela_produtos = pd.read_csv("1-datasets/Produtos.csv", sep=";")
tabela_vendas = pd.read_csv("1-datasets/Vendas.csv", sep=";")
```

```python
# Alterando o nome das colunas
tabela_clientes.columns = ['id_cliente', 'nome', 'segmento', 'cidade', 'estado', 'pais', 'mercado', 'regiao']
tabela_clientes.head()
```

### Conectando o MySQL

```python
# Criando a conexão com MySQL
# Padrão para inserir o caminho: mysql+mysqlconnector://usuario:senha@host/nome_do_banco
engine = create_engine('mysql+mysqlconnector://root:*******/empresa')
```

```python
# Enviando os dados
tabela_vendas.to_sql('vendas', con=engine, if_exists='replace', index=False)
tabela_pedidos.to_sql('pedidos', con=engine, if_exists='replace', index=False)
tabela_produtos.to_sql('produtos', con=engine, if_exists='replace', index=False)
tabela_clientes.to_sql('clientes', con=engine, if_exists='replace', index=False)

print(f"Sucesso! {len(tabela_clientes)} linhas inseridas no MySQL mantendo a estrutura original.")
print(f"Sucesso! {len(tabela_pedidos)} linhas inseridas no MySQL mantendo a estrutura original.")
print(f"Sucesso! {len(tabela_produtos)} linhas inseridas no MySQL mantendo a estrutura original.")
print(f"Sucesso! {len(tabela_vendas)} linhas inseridas no MySQL mantendo a estrutura original.")
```
